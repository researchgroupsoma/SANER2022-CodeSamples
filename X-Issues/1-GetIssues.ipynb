```python
from github import Github
import pandas as pd
```

```python
g = Github("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", per_page=100, timeout=30)
```

```python
frameworks = frameworks = ["googlesamples", "aws-samples", "Azure-Samples", "spring-guides", "googlearchive", "spring-cloud-samples", "SAP-samples"]
```

```python
csvs = list()
for framework in frameworks:
    csvs.append(pd.read_csv(f"../0-GetCodeSamples/{framework}.csv"))
```

```python
samples = pd.concat(csvs, ignore_index=True)
```

```python
len(samples)
```

```python
github_labels = ["bug", "documentation", "duplicate", "enhancement", "good first issue", "help wanted", "invalid", "question", "wontfix"]
```

```python
valid_issues = pd.DataFrame(columns=["sample"])
for sample in samples["full_name"]:
    repo = g.get_repo(sample)
    issues = repo.get_issues(state="all")
    for issue in issues:
        if ("/issues/" in issue.html_url):
            for label in issue.labels:
                if label.name.lower() in github_labels:
                    print(label)
                    label = "question" if "question" == label.name.lower() else "non-question"
                    valid_issues = valid_issues.append({"framwork":sample.split("/")[0],"sample":sample, "body": issue.body, "closed_at": issue.closed_at, "created_at": issue.created_at, "html_url": issue.html_url, "number": issue.number, "state": issue.state, "title": issue.title, "updated_at": issue.updated_at, "label":label}, ignore_index=True)
    print(sample)
    print(g.get_rate_limit())
```

```python
valid_issues.to_csv("issues.csv", index=False)
```
